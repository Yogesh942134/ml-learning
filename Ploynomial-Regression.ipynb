```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = 6 * np.random.rand(200, 1) - 3
y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)

# y = 0.8x^2 + 0.9x + 2
```

```python
plt.plot(X, y,'b.')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# now apply simple Linear regression 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
print(f'The r2 score is {r2_score(y_test,y_pred)}')
```

```python
plt.plot(X, y,'b.')
plt.plot(X_train,lr.predict(X_train))
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# now implement the ploynomial regression
```

```python
tr = PolynomialFeatures()
tr.fit(X_train)
X_train_tr = tr.transform(X_train)
X_test_tr = tr.transform(X_test)
```

```python
plr = LinearRegression()
plr.fit(X_train_tr,y_train)
y_pred = plr.predict(X_test_tr)
print(f'The r2 score is {r2_score(y_test,y_pred)}')
```

```python
X_new= np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = tr.transform(X_new)
y_new = plr.predict(X_new_poly)
```
