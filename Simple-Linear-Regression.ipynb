```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/placement-lr.csv')
```

```python
df.head()
```

```python
# let's see how our data is distributed
plt.scatter(df['cgpa'],df['package'])
plt.xlabel("Cgpa")
plt.ylabel('Package')
```

```python
# train test split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:1],df.iloc[:,1],test_size = 0.2,random_state = 2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test.iloc[[1]])
```

```python
# lets see how our model is predicting , it is acutally drawing a line 
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color = 'red')
plt.xlabel("Cgpa")
plt.ylabel('Package')
```

```python
# behind this , y = mx + b
m = lr.coef_
m
# this is m 
```

```python
b = lr.intercept_ # this is b
b
```

```python
# so , how it predict
m * 8.2 + b
```
