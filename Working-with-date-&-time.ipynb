# Hadle the date

```python
# firstly we have convert date column in to date time data type
date['date'] =  pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# for month
date['month'] =  date['date'].dt.month
date.sample(3)
```

```python
# for year
date['year'] = date['date'].dt.year
date.sample(3)
```

```python
# for days
date['day'] = date['date'].dt.day
date.sample(3)
```

```python
# for month name
date['month_name'] = date['date'].dt.month_name()
date.sample(3)
```

```python
# day of week
date['dayofweek'] = date['date'].dt.dayofweek
date.sample(3)
```

```python
# day week name
date['week-name'] = date['date'].dt.day_name()
date.sample(3)
```

```python
# is weekend ?
date['is-weekend'] = np.where(date['week-name'].isin(['Sunday','Saturday']),'Yes','No')
date.sample(3)
```

```python
# Extract week of the year
date['Week-year'] = date['date'].dt.isocalendar().week
date.sample(3)
```

```python
date['quarter'] = date['date'].dt.quarter
date.sample(3)
```

```python
# for semester
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
date.sample(3)
```

```python
import datetime
today = datetime.datetime.today()
today
```

```python
# difference 
today - date['date']
```

```python
# how manys days only
((today - date['date']).dt.days)
```

```python
# how many months passed
((today - date['date']).dt.days)/31
```
