# we have separete the numerical and categorical data in column number

Link of Pandas 

[Pandas to_numeric()](https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html)

```python
df["number_numeric"] = pd.to_numeric(df['number'],errors = 'coerce',downcast = 'integer')
# errors = 'coerce' if data is not numeric then it convert in to Nan values
# downcast = 'integer' to convert data in to numeric
```

```python
df['number_categorical'] = np.where(df['number_numeric'].isnull(),df['number'],np.nan)
# df['number_numeric'].isnull() is a condition where we want to put values 
# df['number']  for if condition is satisfied then replace with this value of df['number']
# np.nan if condition false fill with nan values
```
