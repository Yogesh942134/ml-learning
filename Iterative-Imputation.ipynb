```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer 
from sklearn.impute import IterativeImputer,KNNImputer,SimpleImputer
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('data/50_Startups.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# iterative imputation
X_train,X_test,y_train,y_test = train_test_split(df[['R&D Spend','Administration']],df['Profit'],test_size=0.2,random_state=42)

itim = IterativeImputer(max_iter=100)
X_train_tr = itim.fit_transform(X_train)
X_test_tr = itim.transform(X_test)

LR = LinearRegression()
LR.fit(X_train_tr,y_train)

y_pred = LR.predict(X_test_tr)
#print('Accuracy is : ',accuracy_score(y_test,y_pred)) it is used for classifiaction problem
print("Cross Value Score : ",cross_val_score(LR,X_train_tr,y_train,cv=10,scoring='r2').mean())
```

```python
# simple imputation
X_train,X_test,y_train,y_test = train_test_split(df[['R&D Spend','Administration']],df['Profit'],test_size=0.2,random_state=42)

itim = SimpleImputer()
X_train_tr = itim.fit_transform(X_train)
X_test_tr = itim.transform(X_test)

LR = LinearRegression()
LR.fit(X_train_tr,y_train)

y_pred = LR.predict(X_test_tr)
#print('Accuracy is : ',accuracy_score(y_test,y_pred)) it is used for classifiaction problem
print("Cross Value Score : ",cross_val_score(LR,X_train_tr,y_train,cv=10,scoring='r2').mean())
```

```python
# knn imputation
X_train,X_test,y_train,y_test = train_test_split(df[['R&D Spend','Administration']],df['Profit'],test_size=0.2,random_state=42)

itim = KNNImputer()
X_train_tr = itim.fit_transform(X_train)
X_test_tr = itim.transform(X_test)

LR = LinearRegression()
LR.fit(X_train_tr,y_train)

y_pred = LR.predict(X_test_tr)
#print('Accuracy is : ',accuracy_score(y_test,y_pred)) it is used for classifiaction problem
print("Cross Value Score : ",cross_val_score(LR,X_train_tr,y_train,cv=10,scoring='r2').mean())
```
