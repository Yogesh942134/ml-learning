```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import mysql.connector #importing my sql connector
```

```python
#giving all necessary info -- syntax
mysqldb = mysql.connector.connect(host='localhost',user='root',password='Yogesh@90',database='campusx')

# creating object cursor to execute the query
var = mysqldb.cursor() 
```

```python
# executing the query and store in var
var.execute("select brand_name,avg(price) from smartphones group by brand_name")

#fetching all data from var and store in result
result = var.fetchall() 

#convert data into dataframe
df = pd.DataFrame(result)
```

```python
df[1] = df[1].astype('int')
```

```python
plt.figure(figsize=(30,20))
plt.bar(df[0],df[1])
plt.xlabel('Smartphones')
```
