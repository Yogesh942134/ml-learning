```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc=50,scale=5,size=10000)
```

```python
sample = pd.Series(data)
```

```python
pd.Series(data).plot(kind='hist')
```

```python
sample_mean = sample.mean()
sample_std = sample.std()
```

```python
from scipy.stats import norm
dist = norm(sample_mean,sample_std)
```

```python
value = np.linspace(sample.min(),sample.max(),100)
```

```python
probabilites = [dist.pdf(value) for value in value]
```

```python
plt.hist(sample,density=True)
plt.plot(value,probabilites)
```

```python
import seaborn as sns
sns.displot(sample)
```

```python
sns.kdeplot(sample)
```

```python
plt.plot(value,probabilites)
```
