```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/batsman_season_record.csv")
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'],bottom=df['2015'])
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'])
```

```python
df.plot(kind='bar')
```

```python
plt.pie(df['2015'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0],shadow=True)
plt.show()
```
