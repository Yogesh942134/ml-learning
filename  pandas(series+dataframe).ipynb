17)apply() (Series/dataframe)            
it can apply function in row wise in data frame and it also work in series as column wise

```python
temp = pd.DataFrame({
    "1st_point" : [(3,4),(4,6),(7,8),(5,6)],
    "2nd_point" : [(-5,6),(9,7),(5,3),(9,1)]
})          
```

```python
def distance(row):
    p1 = row["1st_point"]
    p2 = row["2nd_point"]
    return ((p1[0]-p2[0])**2 - (p1[1] - p2[1])**2)**2
temp.apply(distance,axis=1)
```
