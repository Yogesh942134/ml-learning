**Annotations**

annotation means naming the points on graph

```python
batter = pd.read_csv("data/batter.csv")
```

```python
plt.figure(figsize=(15,7)) # it increase the size of graph
sample = batter.head(25)

plt.scatter(sample["avg"],sample["strike_rate"],s=sample["runs"])
# s for increase the size of dot by how many runs they made
for i in range(sample["avg"].shape[0]):
    plt.text(sample["avg"].values[i],sample["strike_rate"].values[i],sample["batter"].values[i])

# first part x then y and then text
# plt.text is for naming the dot
```

**horizontal and vertical line**

```python
plt.figure(figsize=(15,7)) # it increase the size of graph
sample = batter.head(25)

plt.axhline(130,color="red") # it draw horizontal line 
plt.axvline(30,color="red")  # it draw vertical line 

plt.scatter(sample["avg"],sample["strike_rate"],s=sample["runs"])
# s for increase the size of dot by how many runs they made
for i in range(sample["avg"].shape[0]):
    plt.text(sample["avg"].values[i],sample["strike_rate"].values[i],sample["batter"].values[i])
```

# Sub Plot

```python
plt.style.use("default")
fig,ax = plt.subplots(nrows=2,ncols=2)
# by this syntax we can create multiple graph in one fig

ax[0,0].scatter(sample["avg"],sample["strike_rate"])
ax[0,1].bar(sample["avg"],sample["strike_rate"])
ax[1,0].pie(sample["avg"])
ax[1,1].hist(sample["avg"])

# here if we have to give lable , title or any thing 
# just use set_ before our function

ax[0,0].set_title("avg vs strike rate")
plt.show()
```

**3D surface plot**

```python
ax = plt.subplot(projection='3d')
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

x,y = np.meshgrid(x,y) # meshgrid function help us to create 2D array

z = x**2 + y**2 #this is our z is an equation 
p = ax.plot_surface(x,y,z,cmap="viridis") # syntax for surface plot
plt.colorbar(p) # it just change in syntax for this only
```

**Contour Graph**

```python
#  it convert surface plot in to 2D plot
ax = plt.subplot()
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

x,y = np.meshgrid(x,y) # meshgrid function help us to create 2D array

z = x**2 + y**2 #this is our z is an equation 
p = ax.contour(x,y,z,cmap="viridis") # syntax for contour plot
plt.colorbar(p) 
```

**Contour Fill Graph**

```python
ax = plt.subplot()
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

x,y = np.meshgrid(x,y) # meshgrid function help us to create 2D array

z = x**2 + y**2 #this is our z is an equation 
p = ax.contourf(x,y,z,cmap="viridis") # syntax for contour fill plot
plt.colorbar(p) 
```

**heatmap**

```python
delivery = pd.read_csv("data/IPL_Ball_by_Ball_2008_2022.csv")
a = delivery[(delivery["ballnumber"].isin([1,2,3,4,5,6])) & (delivery["batsman_run"] == 6)]

b = a.pivot_table(index="overs",columns="ballnumber",values="batsman_run",aggfunc="count")
# for plot heatmap we need grid fromat data so above process make data in grid form
plt.imshow(b) # syntax for heatmap
plt.colorbar()
```
