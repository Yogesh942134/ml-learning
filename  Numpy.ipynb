**Broadcasting**

#Rules

1) if the number of dimensions of the two array is not same then add new dimension with size 1 to the head of array with smaller dimensions
2) if the size of each dimension does not match then strech then dimenstion with size 1 are streched to the size of the other array
3) if there dimension whose size is not 1 in either of two array , it can not be broadcasted and an error is raised

```python
#example 
a = np.arange(12).reshape(4,3)
b = np.arange(3).reshape(3)
# here we add 1 to head os b array (1,3) and 1 is streched up to 4 then they are same thats why it work
# (4,3) = (4,3)
a+b
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3).reshape(3) 
# here we add 1 to head os b array (1,3) and 1 is streched up to 3 but they donot match thats why here it dont work
# (3,4) != (3,3)
#a+b
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)
#(3,3) = (3,3)
a+b
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)
# (4,3) = (4,3)
a*b # we can perform any operation here
```

```python
a = np.arange(12).reshape(4,3)
b = np.arange(12).reshape(3,4)
#(4,3) != (3,4) #here 1 is not present rule 3 is apply
#a+b
```

#set fuction

```python
"""
1)np.union1d = it gives union of two array
2)np.intersect1d = it gives intrescetion of array
3)np.setdiff1d = np.setdiff[a1,a2] it gives elements only present in a1 i.e in only first one array
4)np.setxor1d = it gives elements which are not common i.e it do no give elements who are present in intersection
"""
a = np.random.randint(1,100,10)
b = np.random.randint(1,100,10)
print(a)
print(b)
print(np.union1d(a,b))
print(np.intersect1d(a,b))
print(np.setdiff1d(a,b))
print(np.setxor1d(a,b))
```
