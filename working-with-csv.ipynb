# 3. Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv" # here we have to just paste the link 
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)

# this is code snippet
```

# 13. Handling Dates

```python
pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info() # by this we change a object to date time 
```

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

# 14. Convertors

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename}).head(5) # by this we can change the any information by passing function 
```
