```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
df = pd.read_csv('data/train.csv',usecols=['Age','SibSp','Parch','Fare','Survived'])
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
# let's check first cross val score before feature extraction
X = df.iloc[:,1:]
y = df.iloc[:,0]

print('Cross Val Score : ',cross_val_score(LogisticRegression(),X,y,cv=20,scoring='accuracy').mean())
```

# Feature Extraction -> Creating New Col using Old 

```python
df['Family_size'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.head()
```

```python
# let's create a function to divide this family in to small,medium and large
def size_f(num):
    if num == 1:
        return 0 # small
    elif num > 1 and num <= 4:
        return 1 # medium 
    else:
        return 2 # large
```

```python
size_f(7)
```

```python
df['Family_size'] = df['Family_size'].apply(size_f)
```

```python
df.head()
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]

print('Cross Val Score : ',cross_val_score(LogisticRegression(),X,y,cv=20,scoring='accuracy').mean())
```

# Feature Splitting 

```python
df = pd.read_csv('data/train.csv')[['Name','Survived']]
```

```python
# this is logic for extracting title
df['Title'] = df['Name'].str.split(',',expand=True)[1].str.split('.',expand=True)[0]
```

```python
# now we can perform analysis
df[(df['Survived'] == 0)]['Title'].value_counts()/len(df)
```
