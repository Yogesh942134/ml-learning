```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('data/iris.csv')
```

```python
df.head()
```

```python
df = df[['SepalWidthCm','PetalWidthCm','Species']]
```

```python
df.sample(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
```

```python
df.sample(3)
```

```python
# this is now problem binary classification
df = df[df['Species'] > 0]
```

```python
plt.scatter(x=df['SepalWidthCm'],y=df['PetalWidthCm'],c=df['Species'],cmap='winter')
```

```python
# now makke data sets for dicision tree 
```

```python
df = df.sample(100)
df_train = df.iloc[:60,:].sample(10) # testing
df_val = df.iloc[60:80,:].sample(5) # for valdition
df_test = df.iloc[80:,:].sample(5) # final combine tesing
```

```python
# make data for testing each decision tree
X_test = df_val.iloc[:,0:2]
y_test = df_val.iloc[:,-1]
```

```python
def Evaluate(clf,X,y):
    clf.fit(X,y)
    plot_tree(clf)
    plt.show()
    plot_decision_regions(X.values, y.values, clf=clf, legend=2)
    y_pred = clf.predict(X_test)
    print(f'Accuracy is {accuracy_score(y_test,y_pred)}')
```

```python
# for decision tree 1
X_train = df_train.iloc[:,:2].sample(10,replace=True)
y_train = df_train.iloc[:,-1].sample(10,replace=True)
clf1 = DecisionTreeClassifier()
Evaluate(clf1,X_train,y_train)
```

```python
# for decision tree 2
X_train = df_train.iloc[:,:2].sample(10,replace=True)
y_train = df_train.iloc[:,-1].sample(10,replace=True)
clf2 = DecisionTreeClassifier()
Evaluate(clf2,X_train,y_train)
```

```python
# for tree 3
X_train = df_train.iloc[:,:2].sample(10,replace=True)
y_train = df_train.iloc[:,-1].sample(10,replace=True)
clf3 = DecisionTreeClassifier()
Evaluate(clf3,X_train,y_train)
```
